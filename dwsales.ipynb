```python
# we can see that in amounts collumn we have 12 null values
# now we have two options either we can fill those null values or just delete 
df.dropna(inplace = True)    # in basic word inplace = true means save it thats it
```

```python
# Exploratory Data Analysis  EDA
```

```python
# we want to plot a bar chart with label to know no of males and females
a = sns.countplot(x = 'Sex', data = df)

for bars in a.containers:
    a.bar_label(bars)
```

```python
sale_by_Gen = df.groupby(['Sex'], as_index=False)['Amount'].sum().sort_values(by ='Amount', ascending=False)
```

```python
sale_by_Gen
```

```python
sns.barplot(x= 'Sex', y= 'Amount', data = sale_by_Gen)
```

```python
# from the above graph we can say that most of the buyers are female and females have the more purchasing power then male
```

```python
p = sns.countplot(data = df, x = "Age Group", hue= 'Sex')
```

```python
# for labelling bars
p = sns.countplot(data = df, x = "Age Group", hue= 'Sex')    # here hue = 'Sex' will be the legend

for bars in p.containers:
    p.bar_label(bars)
```

```python
# from the above graph we can say that most of the buyers are of age group 26 to 35 years females 
```

```python
sale_by_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.barplot(x = 'Age Group', y = 'Amount', data = sale_by_age)
```

```python
# total no of orders from top 10 states
sale_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,6)})     # chart size
sns.barplot(data = sale_state, x = 'State', y = "Orders")
```

```python
# total no of Amount from top 10 states
sale_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,6)})     # chart size
sns.barplot(data = sale_state, x = 'State', y = "Amount")
```

```python
# from the above graph we can say that most of the orders and amount spent states are UP, Maharastra, Karnataka
```

```python
# shopping based on the maritial status
ax = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(2,1)})     # chart size
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(20,8)})     # chart size
ax = sns.countplot(data = df, x= "Occupation")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
pp_sector = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.barplot(data = pp_sector, x = 'Occupation', y = 'Amount')
```

```python

# from the above graph we can say that the It, health, avation are the top 3 most buyer sector 


#CONCLUSION
#Married Womens of age group 26-35yrs from UP, Maharastra and Karnataka working in It, Healthcare and Avaition are more likely to buy products from food, 
#clothings and electronics category
```
